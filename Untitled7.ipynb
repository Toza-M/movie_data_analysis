```python
import pandas as pd
import numpy as np

df = pd.read_csv("movies_dataset_final.csv")
df.head()
```

```python
clean = df.copy()
```

```python
money_columns = ['Budget', 'Domestic Gross', 'Worldwide Gross']

for col in money_columns:
    clean[col] = clean[col].replace('[\$,]', '', regex=True).astype(float)
```

```python
clean['Release Date'] = pd.to_datetime(clean['Release Date'])
```

```python
clean['Season'] = clean['Season'].astype('category')
```

```python
clean.isnull().sum()
```

```python
clean = clean.dropna()    # we didn't use it because there is no missing data
```

```python
clean = clean.fillna(clean.mean(numeric_only=True))# we didn't use it because there is no missing data to fill
```
